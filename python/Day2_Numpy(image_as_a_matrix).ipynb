```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from google.colab import files

uploaded = files.upload()

image = mpimg.imread('sample.jpg')

print("Image Shape:", image.shape)

# Display Original Image
plt.figure(figsize=(5,5))
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()

# STEP 3: Convert to Grayscale
gray_image = np.mean(image, axis=2)

# STEP 4: Increase Brightness
bright_image = image + 50

# Keep values between 0 and 255
bright_image = np.clip(bright_image, 0, 255)

# Convert datatype
bright_image = bright_image.astype(np.uint8)

# STEP 5: Decrease Brightness
dark_image = image - 50
dark_image = np.clip(dark_image, 0, 255)
dark_image = dark_image.astype(np.uint8)

# STEP 6: Flip Image Horizontally
horizontal_flip = np.fliplr(image)

# STEP 7: Flip Image Vertically
vertical_flip = np.flipud(image)

# STEP 8: Crop Image
cropped_image = image[100:400, 200:500]

# STEP 9: Blur Effect (BONUS)
blur_image = image.copy()

for i in range(1, image.shape[0]-1):
    for j in range(1, image.shape[1]-1):

        blur_image[i, j] = np.mean(
            image[i-1:i+2, j-1:j+2],
            axis=(0,1)
        )

blur_image = blur_image.astype(np.uint8)

# STEP 10: Display All Images
fig, axes = plt.subplots(2, 4, figsize=(18,10))

# Original
axes[0,0].imshow(image)
axes[0,0].set_title("Original")

# Grayscale
axes[0,1].imshow(gray_image, cmap='gray')
axes[0,1].set_title("Grayscale")

# Bright
axes[0,2].imshow(bright_image)
axes[0,2].set_title("Brightened")

# Dark
axes[0,3].imshow(dark_image)
axes[0,3].set_title("Darkened")

# Horizontal Flip
axes[1,0].imshow(horizontal_flip)
axes[1,0].set_title("Horizontal Flip")

# Vertical Flip
axes[1,1].imshow(vertical_flip)
axes[1,1].set_title("Vertical Flip")

# Cropped
axes[1,2].imshow(cropped_image)
axes[1,2].set_title("Cropped")

# Blur
axes[1,3].imshow(blur_image)
axes[1,3].set_title("Blur Effect")

# Remove axis
for ax in axes.flat:
    ax.axis('off')

plt.tight_layout()
plt.show()

```
