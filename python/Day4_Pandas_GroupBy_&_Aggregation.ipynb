```python
# =========================================
# DAY 4 — SALES PERFORMANCE DASHBOARD
# ALTERNATIVE GOOGLE COLAB CODE
# =========================================

# =========================================
# STEP 1: IMPORT LIBRARIES
# =========================================

import pandas as pd
import zipfile
import os

# =========================================
# STEP 2: UPLOAD ZIP FILE
# =========================================

from google.colab import files

uploaded = files.upload()

# =========================================
# STEP 3: FIND ZIP FILE AUTOMATICALLY
# =========================================

zip_file = None

for file in os.listdir():
    if file.endswith(".zip"):
        zip_file = file
        break

print("ZIP File Found:", zip_file)

# =========================================
# STEP 4: EXTRACT ZIP FILE
# =========================================

with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall()

print("ZIP Extracted Successfully ✅")

# =========================================
# STEP 5: FIND CSV FILE AUTOMATICALLY
# =========================================

csv_file = None

for file in os.listdir():
    if file.endswith(".csv"):
        csv_file = file
        break

print("CSV File Found:", csv_file)

# =========================================
# STEP 6: LOAD DATASET
# =========================================

df = pd.read_csv(csv_file, encoding='latin1')

# =========================================
# STEP 7: LOWERCASE COLUMN NAMES
# =========================================

df.columns = df.columns.str.lower()

print("\nColumns:")
print(df.columns)

# =========================================
# STEP 8: FIRST 5 ROWS
# =========================================

print("\nFirst 5 Rows:")
print(df.head())

# =========================================
# STEP 9: DATASET INFO
# =========================================

print("\nDataset Info:")
print(df.info())

# =========================================
# STEP 10: DESCRIBE DATA
# =========================================

print("\nDataset Statistics:")
print(df.describe())

# =========================================
# STEP 11: TOTAL SALES & PROFIT
# BY REGION AND CATEGORY
# =========================================

print("\nSales and Profit by Region & Category:")

sales_profit = df.groupby(
    ['region', 'category']
)[['sales', 'profit']].sum()

print(sales_profit)

# =========================================
# STEP 12: TOP 3 SUB-CATEGORIES
# =========================================

print("\nTop 3 Sub-Categories in Each Region:")

top_sub = df.groupby(
    ['region', 'sub-category']
)['sales'].sum().reset_index()

top_sub = top_sub.sort_values(
    ['region', 'sales'],
    ascending=[True, False]
)

top3 = top_sub.groupby('region').head(3)

print(top3)

# =========================================
# STEP 13: PROFIT MARGIN COLUMN
# =========================================

df['profit_margin_%'] = (
    df['profit'] / df['sales']
) * 100

print("\nProfit Margin Added ✅")

print(df[['sales', 'profit', 'profit_margin_%']].head())

# =========================================
# STEP 14: PIVOT TABLE
# =========================================

print("\nPivot Table:")

pivot = pd.pivot_table(
    df,
    values='sales',
    index='region',
    columns='segment',
    aggfunc='mean'
)

print(pivot)

# =========================================
# STEP 15: NEGATIVE PROFIT MONTHS
# =========================================

df['order date'] = pd.to_datetime(df['order date'])

df['month'] = df['order date'].dt.to_period('M')

monthly_profit = df.groupby('month')['profit'].sum()

negative_months = monthly_profit[
    monthly_profit < 0
]

print("\nMonths with Negative Profit:")

print(negative_months)

# =========================================
# STEP 16: APPLY + LAMBDA
# =========================================

df['performance'] = df['sales'].apply(
    lambda x: 'High Sales'
    if x > 500 else 'Low Sales'
)

print("\nPerformance Column:")

print(df[['sales', 'performance']].head())

# =========================================
# BONUS CHALLENGE
# QUARTERLY SALES TREND
# =========================================

quarterly_sales = df.resample(
    'QE',
    on='order date'
)['sales'].sum()

print("\nQuarterly Sales Trend:")

print(quarterly_sales)

# =========================================
# PROJECT COMPLETED
# =========================================

print("\nProject Completed Successfully ✅")
```
